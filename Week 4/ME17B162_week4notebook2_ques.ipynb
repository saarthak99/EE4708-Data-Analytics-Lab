Write a function for doing one hot encoding for all categorical features

Hint: Use pandas.get_dummies

```python
def onehot_encoding(df,columns):
    '''
    Takes the dataframe 
    columns which are corresponding to categorical features
    '''
    for i in range(len(columns)):
        #rated_dummies = pd.get_dummies(df[columns[i]])
        df = pd.concat([df, pd.get_dummies(df[columns[i]])], axis=1)
        del df[columns[i]]
    return df

newdf = onehot_encoding(housedf,cat)

newdf['SalePrice'] = housedf['SalePrice']
```

**Conclusion**
